```python
URL = 'https://openlibrary.org/api/books?'
PARAMS = 'bibkeys=ISBN:{}&jscmd=data&format=json'
```

```python
def get_available_books(books_isbn, n=100):
    res = []
    for b in range(n):
        print("book: {}".format(b))
        if (b %1000) == 0:
            output_list_to_file('batch: {}'.format(b), res)
        local_params = PARAMS.format(books_isbn[b])
        r = requests.get(url = URL, params=PARAMS)
        data = r.json()
        if 'ISBN:{}'.format(books_isbn[b]) not in data:
            continue
        data = data['ISBN:{}'.format(books_isbn[b])]
        if data['ebooks'][0]['availability'] == 'full':
            res.append(books_isbn[b])
    
    return res
```

```python
def output_list_to_file(batchname, l, filename='available_books.txt'):
    with open(filename, 'w') as f:
        f.write(batchname + '\n')
        for row in l:
            f.write(str(row) + '\n')
```
